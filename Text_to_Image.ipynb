```python
def read_numbers_to_matrix(filename,n):
    with open(filename, 'r') as f:
        matrix=[]
        for i in range(n):
            L=[]
            for i in range(n):
                l=int(float(f.readline().strip()))
                L.append(l)
            matrix.append(L)
    return matrix
```

```python
#Grayscale Image
matrix = read_numbers_to_matrix("C:\\Users\\Dell\\OneDrive\\Documents\\college\\hpc\\output_grey.txt",453)
```

```python
import numpy as np
matrix=np.array(matrix)
matrix=matrix.astype(np.uint8)
```

```python
import cv2
```
